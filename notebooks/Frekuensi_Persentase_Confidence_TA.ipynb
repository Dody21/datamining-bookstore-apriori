```python
top10 = df.sort_values(by="Confidence (%)", ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.barh(top10["Itemset"], top10["Confidence (%)"], color="skyblue")
plt.xlabel("Confidence (%)")
plt.title("Top 10 Itemset dengan Confidence Tertinggi")
plt.gca().invert_yaxis()  # Membalik urutan agar yang tertinggi di atas
plt.tight_layout()

# Simpan sebelum show
plt.savefig("top10_confidence_itemset.png", dpi=300, bbox_inches='tight')

# Tampilkan grafik
plt.show()
```
