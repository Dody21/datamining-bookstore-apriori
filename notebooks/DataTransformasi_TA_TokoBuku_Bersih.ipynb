# Import library

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
```

# Load dataset

```python
file_path = 'DataCleaningTA.csv'
df = pd.read_csv(file_path)
```

# Transformasi data

```python
# Mengubah string list menjadi list Python
df['jenis_item'] = df['jenis_item'].apply(eval)

# One-hot encoding terhadap produk
mlb = MultiLabelBinarizer()
produk_encoded = mlb.fit_transform(df['jenis_item'])
produk_df = pd.DataFrame(produk_encoded, columns=mlb.classes_)

# Menambahkan kolom tanggal pembelian dan nomor transaksi
final_df = pd.concat([df['tanggal pembelian'], produk_df], axis=1)
final_df.insert(1, 'transaksi', range(1, len(final_df) + 1))

# Menampilkan seluruh DataFrame
# Pandas default hanya menampilkan 60 baris dan 20 kolom, jadi kita ubah pengaturannya
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Menampilkan seluruh data
display(final_df)
```

# Menyimpan hasil ke file

```python
final_df.to_csv('DataTransformasiTA.csv', index=False)
print("\nDataset transformasi sudah disimpan sebagai 'DataTransformasiTA.csv'")
```
